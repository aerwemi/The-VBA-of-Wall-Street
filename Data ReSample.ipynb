```python
# OUTPUT 1st and last raw of each stock
writer = pd.ExcelWriter('Multiple_year_stock_closeStartEnd2.xlsx', engine='xlsxwriter')
for i in range(2012,2017,1):
    df=pd.read_excel('Multiple_year_stock_data.xlsx', sheetname=str(i))
    print(df.shape)
    
    #### 
    df = df[df['<vol>'] != 0]
    
    StockSet = set(df.sort_values('<ticker>')['<ticker>'].tolist())
    StockL=[]
    for Stock in StockSet:
        dfStock=df[df['<ticker>'] == Stock]
        dfStock=dfStock.iloc[[0, -1]]
        if dfStock.shape[0] > 1:
            dfStock.sort_values(['<ticker>', '<date>'], inplace=True)
            StockL.append(dfStock)
            
    
    df_concat=pd.concat(StockL)
    df_concat.sort_values(['<ticker>', '<date>'], inplace=True)
    print(df_concat.head())
    
    ####
    df_concat.to_excel(writer, index=False, sheet_name=str(i))

    print(df.shape)
    print(df_concat.shape)
    
    print('----')
    
writer.save()
```
